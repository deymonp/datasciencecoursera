### Sampling data
- looping through a file and retrieving only necessary lines of code

```python
#retriving even rows from the file (looping)
n = 2 #setting up a denominator to recieve an even number
with open("Colors.txt", 'r') as open_file:
    for j, observation in enumerate(open_file):
        if j%n == 0:
            print("Read line: " + str(j) + " " + "Content: " + observation, end="")
```

```python
# random sampling
sample_size = 0.25
from random import random
with open("Colors.txt", 'r') as open_file:
    for j, observation in enumerate(open_file):
        if random() <= sample_size:
            print("Read line: " +  str(j) + " " + "Content" + observation, end="")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the page containing Tesla revenue data
url = "https://example.com/tesla-revenue"  # Replace with the actual URL

# Send a GET request to the URL
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Print the entire HTML content for debugging (optional)
# print(soup.prettify())  # Uncomment for debugging

# Find the table containing revenue data
table = soup.find('table')  # Adjust or specify the class if necessary

# Check if the table was found
if table is None:
    print("Table not found. Please check the URL and the table selector.")
else:
    # Parse the table rows
    data = []
    for row in table.find_all('tr')[1:]:  # Skip the header row
        cols = row.find_all('td')
        data.append([col.text.strip() for col in cols])

    # Create a DataFrame
    tesla_revenue = pd.DataFrame(data, columns=['Year', 'Revenue'])

    # Display the first five rows
    print(tesla_revenue.head())
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

def download_tesla_data(start_date, end_date):
    """Download Tesla stock data from Yahoo Finance."""
    tesla_data = yf.download('TSLA', start=start_date, end=end_date)
    return tesla_data

def save_to_csv(dataframe, filename):
    """Save DataFrame to a CSV file."""
    dataframe.to_csv(filename, index=False)

def plot_tesla_data(dataframe):
    """Plot Tesla stock closing price."""
    plt.figure(figsize=(10, 5))
    plt.plot(dataframe['Date'], dataframe['Close'], label='Closing Price', color='blue')
    plt.title('Tesla Stock Price (TSLA)')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.grid()
    plt.show()

# Main script
if __name__ == "__main__":
    # Get Tesla stock data
    tesla_data = download_tesla_data('2023-01-01', '2024-10-19')

    # Reset index, save, and display the first five rows
    tesla_data.reset_index(inplace=True)
    save_to_csv(tesla_data, 'tesla_stock_data.csv')
    print(tesla_data.head())

    # Plot the Tesla stock data
    plot_tesla_data(tesla_data)
```
